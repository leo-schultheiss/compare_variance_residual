```python
# https://github.com/garland-culbreth/nord-matplotlib?tab=readme-ov-file

import matplotlib

matplotlib_dir = matplotlib.get_configdir()
matplotlib_dir
```

```python
import os

subdirs = os.listdir(matplotlib_dir)
os.makedirs(os.path.join(matplotlib_dir, "stylelib"), exist_ok=True)
stylelib_dir = os.path.join(matplotlib_dir, "stylelib")
stylelib_dir
```

```python
nord_theme_names = [ "nord.mplstyle","nord-talk.mplstyle", "nord-dark-talk.mplstyle", "nord-dark.mplstyle", "nord-light-talk.mplstyle", "nord-light.mplstyle", ]
nord_theme = nord_theme_names[0]
```

Download theme

```python
import requests

url = "https://raw.githubusercontent.com/garland-culbreth/nord-matplotlib/main/nord.mplstyle"
response = requests.get(url)
if response.status_code == 200:
    with open(os.path.join(stylelib_dir, nord_theme), "wb") as f:
        f.write(response.content)
        print("Theme downloaded successfully.")
else:
    print(f"Failed to download theme: {response.status_code}")
```

```python
import matplotlib.pyplot as plt
plt.style.use("nord")
```

```python
X = [1,2,3,4,5]
Y = [1,2,3,4,5]
plt.plot(X,Y)
plt.show()
```
