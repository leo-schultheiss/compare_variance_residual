```python
import os

import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import simplstyles
from results import get_result_path
```

```python
plt.style.use('nord-light-talk')
```

```python
subject = 5
modality = 'reading'
low_level_feature = 'letters'
```

```python
vp_correlation = pd.read_csv(os.path.join(get_result_path(modality, subject), f"vp_english1000_{low_level_feature}_scores.csv"))[f"semantic\\{low_level_feature}"]
rm_correlation = pd.read_csv(os.path.join(get_result_path(modality, subject), f"residual_{low_level_feature}_english1000_scores.csv"))[f"correlation_score"]
```

```python
fig, ax = plt.subplots(1,1, figsize = (4,4))
max_total = max(vp_correlation.max(), rm_correlation.max())
min_total = min(vp_correlation.min(), rm_correlation.min())
sns.scatterplot(x=vp_correlation, y=rm_correlation, alpha=0.1)
# sns.kdeplot(x=vp_correlation, y=rm_correlation)

plt.plot([min_total, max_total], [min_total, max_total], color='C2', linestyle='--', label='x=y')
plt.xlabel("Variance Partitioning")
plt.ylabel("Residual Method")
plt.ylim(min_total, max_total)
plt.xlim(min_total, max_total)
```
