```python
import numpy as np
from scipy.stats import zscore
```

```python
n_targets = 5
n_features = 10
n_samples = 100
```

```python
A, B, C = np.random.randn(n_samples, n_features), np.random.randn(n_samples, n_features), np.random.randn(n_samples, n_features)
A, B, C = zscore(A, axis=0), zscore(B, axis=0), zscore(C, axis=0)
```

```python
B_A, B_B, B_C = np.random.randn(n_features, n_targets), np.random.randn(n_features, n_targets), np.random.randn(n_features, n_targets)
B_A, B_B, B_C = zscore(B_A, axis=0), zscore(B_B, axis=0), zscore(B_C, axis=0)
```

```python
X_1, X_2 = np.hstack([A, B]), np.hstack([A, C])
X_1, X_2 = zscore(X_1, axis=0), zscore(X_2, axis=0)
```

```python
B_1, B_2 = np.vstack([B_A, B_B]), np.vstack([B_A, B_C])
B_1, B_2 = zscore(B_1, axis=0), zscore(B_2, axis=0)
```

```python
a_1, a_2 = 1, 1
Y = a_1 * X_1 @ B_1 + a_2 * X_2 @ B_2
Y= zscore(Y, axis=0)
Y
```
